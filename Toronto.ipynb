#### Reading the table in the webpage 

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url,header=0)
df=df[0]
```

#### Updated dataframe with latitude & longitude

```python
data['Borough']=data.Borough.str.extract(r"(.*Toronto)",expand=False)
data.head()
```

```python
data.dropna(subset=["Borough"], axis=0, inplace=True)
data.reset_index(drop=True,inplace=True)
data.head()
```

```python
address='Toronto,Ontario'
geolocator=Nominatim(user_agent='toronto_explorer')
location=geolocator.geocode(address)
latitude=location.latitude
longitude=location.longitude
print('{} is the latitude and longitude of Toronto'.format(latitude,longitude))

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(data['Latitude'], data['Longitude'], data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
